## Sex column의 값을 int형으로 바꿔주기

값이 str형이면 코드가 제대로 실행이 되지 않기 때문이다.  
그렇기 때문에 간단히 male을 0으로, female을 1로 바꿔줄 것이다.

```python
X['Sex'].head(10)
```

```python
y['Sex'].head(10)
```

```python
#str형에서 int형으로 변경
X['Sex'] = X['Sex'].replace(['male', 'female'], [0, 1])
y['Sex'] = y['Sex'].replace(['male', 'female'], [0, 1])
```

```python
X.head(10)
```

```python
y.head(10)
```

Sex column의 값들이 male, female에서 각각 0, 1로 바뀐 것을 확인할 수 있다.
